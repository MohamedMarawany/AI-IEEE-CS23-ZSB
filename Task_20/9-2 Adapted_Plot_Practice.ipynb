```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

In this workspace, you will work with the fuel economy dataset from the previous lesson on bivariate plots.

```python
fuel_econ = pd.read_csv('fuel-econ.csv')
fuel_econ.head()
```

**Task 1**: Plot the city ('city') vs. highway ('highway') fuel efficiencies (both in mpg) for each vehicle class ('VClass'). Don't forget that vehicle class is an ordinal variable with levels {Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, Large Cars}.

```python
fuel_econ['VClass'].unique()
```

```python
# Define Vehicles Classes
vehicle_class = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

# Create Plots
plt.figure(figsize=(10, 6))
sb.scatterplot(data=fuel_econ, x='city', y='highway', hue='VClass' ,hue_order=vehicle_class)

# Set axis Labels
plt.xlabel('City Fuel Efficiency (mpg)')
plt.ylabel('Highway Fuel Efficiency (mpg)')
plt.title('City vs. Highway Fuel Efficiencies by Vehicle Class')

# Add a legend
plt.legend(title='Vehicle Class', title_fontsize='12')
plt.show()
```

**Task 2**: Plot the relationship between engine size ('displ', in liters), vehicle class, and fuel type ('fuelType'). For the lattermost feature, focus only on Premium Gasoline and Regular Gasoline cars. What kind of relationships can you spot in this plot?

```python
# Filter the DataFrame to include only Premium Gasoline and Regular Gasoline cars
fuel_types_to_plot = ['Premium Gasoline', 'Regular Gasoline']
filtered_fuel_econ = fuel_econ[fuel_econ['fuelType'].isin(fuel_types_to_plot)]

sb.pointplot(data = filtered_fuel_econ, x = 'VClass', y = 'displ', hue = 'fuelType',
             dodge = 0.3, linestyles = "")

# Set axis labels and title
plt.xlabel('Vehicle Class')
plt.ylabel('Engine Size (liters)')
plt.title('Engine Size vs. Vehicle Class by Fuel Type')

# Add a legend and customize its title
plt.legend(title='Fuel Type', title_fontsize='12')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Filter the DataFrame to include only Premium Gasoline and Regular Gasoline cars
fuel_types_to_plot = ['Premium Gasoline', 'Regular Gasoline']
filtered_fuel_econ = fuel_econ[fuel_econ['fuelType'].isin(fuel_types_to_plot)]

# Define the order of the vehicle classes
vehicle_class_order = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

# Create a swarm plot using seaborn's swarmplot function
plt.figure(figsize=(12, 6))  # Adjust the size of the plot if needed
sb.swarmplot(data=filtered_fuel_econ, x='VClass', y='displ', hue='fuelType', order=vehicle_class_order, palette='Set1')

# Set axis labels and title
plt.xlabel('Vehicle Class')
plt.ylabel('Engine Size (liters)')
plt.title('Engine Size vs. Vehicle Class by Fuel Type')

# Add a legend and customize its title
plt.legend(title='Fuel Type', title_fontsize='12')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```
