# EDA with Visuals
Create visualizations to answer the questions below. Use the `winequality_edited.csv` dataset, which you should've created  in the previous quiz.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('winequality_edited.csv')
df.head()
```

### Which of the following feature variables appear skewed to the right? Fixed Acidity, Total Sulfur Dioxide, pH, Alcohol _____________________ (answer using plots)

```python
plt.hist(df['fixed_acidity'], bins=20, alpha=0.5, label='Fixed Acidity')
plt.hist(df['total_sulfur_dioxide'], bins=20, alpha=0.5, label='Total Sulfur Dioxide')
plt.hist(df['pH'], bins=20, alpha=0.5, label='pH')
plt.hist(df['alcohol'], bins=20, alpha=0.5, label='Alcohol')
plt.legend()
plt.show()
```

### Which of the following is most likely to have a positive impact on quality? Volatile Acidity, Residual Sugar, pH, Alcohol _____________________________ (answer using plots)

```python
sns.scatterplot(x='volatile_acidity', y='quality', data=df)
plt.show()
```

```python
sns.scatterplot(x='residual_sugar', y='quality', data=df)
plt.show()
```

```python
sns.scatterplot(x='pH', y='quality', data=df)
plt.show()
```

```python
sns.scatterplot(x='alcohol', y='quality', data=df)
plt.show()
```
